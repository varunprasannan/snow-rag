```

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```

spark = SparkSession.builder.appName("FI_factInvoice").getOrCreate()

staging_rseg = "abfss://staging@datalake.dfs.core.windows.net/sap/rseg/"
```

```

rseg_df = spark.read.parquet(staging_rseg)
```

```

spark.sql("""
CREATE TABLE IF NOT EXISTS fi.factInvoice (
    BELNR STRING,
    GJAHR STRING,
    MATNR STRING,
    WRBTR DECIMAL(18,2)
)
USING DELTA
""")
```

```

rseg_df.select(
    "BELNR",
    "GJAHR",
    "MATNR",
    "WRBTR"
).write.mode("overwrite").saveAsTable("fi.factInvoice")
```

```

spark.sql("VACUUM fi.factInvoice RETAIN 168 HOURS")
```
