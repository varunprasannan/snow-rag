```

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```

spark = SparkSession.builder.appName("SC_factGoodsReceipt").getOrCreate()

staging_mseg = "abfss://staging@datalake.dfs.core.windows.net/sap/mseg/"
```

```

mseg_df = spark.read.parquet(staging_mseg)
```

```

spark.sql("""
CREATE TABLE IF NOT EXISTS sc.factGoodsReceipt (
    MBLNR STRING,
    MATNR STRING,
    WERKS STRING,
    MENGE DECIMAL(18,2)
)
USING DELTA
""")
```

```

mseg_df.select(
    "MBLNR",
    "MATNR",
    "WERKS",
    "MENGE"
).write.mode("overwrite").saveAsTable("sc.factGoodsReceipt")
```

```

spark.sql("VACUUM sc.factGoodsReceipt RETAIN 168 HOURS")
```
